```r
#Q1 sd_x
x <- c(1,2,3)
sd_x = function(x){sqrt(sum((x-mean(x))^2)/(length(x)-1))}
      
sd_x(x)
```

```r
#Q2
#input coral species in vector
coral_pop = c("Montipora capitata",
              "Porites compressa",
              "Porites lobata")
#sample 100 values from the vector coral_pop and use replace=TRUE to keep the result still
sample_coral = sample(coral_pop,size=100,replace=TRUE)

#input TRUE,FALSE in vector
health_pop = c(TRUE,FALSE)
#sample 100 values from TRUE and FALSE 
#use function prob to set conditional probability of event
health_coral = sample(health_pop,size=100,
                      replace=TRUE,
                      prob=c(0.3,0.8))
# crate dataframe 
coral_tibble = data.frame(sample_coral,health_coral)
coral_tibble
```

```r
#Q3
# create function  
# healthy coral (health_status = TRUE) sample the cover values from a normal with mean 9 and std of 2
# diseased corals (health_status = FALSE), sample the cover values from a normal with mean 1 and std of 3

cover = function(x){
    if (x==TRUE) {
        rnorm(1,9,2)}
    else if (x==FALSE){
        rnorm(1,1,3)}
}
# crate new vector 
coral_cover <- mapply(cover, coral_tibble$health_coral)

# crate new data frame 
coral_tibble_cover = data.frame(sample_coral,health_coral,coral_cover )
coral_tibble_cover
```

```r
#Q4

binary_status <- sapply(coral_tibble$health_coral,as.integer)

coral_tibble_cover_binary = data.frame(sample_coral,health_coral,coral_cover,binary_status)
coral_tibble_cover_binary
```

```r
#Q5
ggplot(coral_tibble_cover_binary, aes(x=coral_cover, ))+
    geom_histogram (aes(y = (..count..)/sum(..count..)), 
    fill="gray", colour="black", binwidth=1) +
    labs(title="distribution of healthy and diseased corals",
    x="x_axes",y="y_axes") +
    facet_grid (health_coral ~ .) 
```

```r
#Q6

ggplot(coral_tibble_cover_binary, aes(x=coral_cover, ))+
    labs(title="coral species distribution",
    x="x_axes",y="y_axes") +
    facet_grid (sample_coral ~ .) +
    geom_density( fill ="green", alpha=0.2 ,adjust = 2) +
    xlim(c(-20,30))
```
